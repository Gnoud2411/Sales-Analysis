# Sales Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb

# Turn off warnings
import warnings
warnings.filterwarnings('ignore')
```

## I. Prepare

```python
# Load dataframe
data = pd.read_feather(r"F:\Sales_ Analysis\Sales_data.ftr")
data.head()
```

## II. Data cleaning
### 1. Drop "Null" values

```python
# Identify the total number of NA values in the dataframe
data.isnull().sum()
```

```python
# Eliminate NA values
drop_na_data = data.dropna(how = "all")
```

### 2. Drop duplicate data

```python
# Identify duplicate values in dataframe drop_na_data
drop_na_data[drop_na_data.duplicated()].head()
```

```python
# Eliminate duplicate values
clean_data = drop_na_data.drop_duplicates()
```

## III. Analyze and Visualize 
### 1. Identify total sales by month

```python
# Create a funtion that returns Month
def return_month(date):
    return date.split('/')[0]
```

```python
# Create a new field containing the Month values
clean_data['Month'] = clean_data['Order Date'].apply(return_month)
```

```python
# Identify data type of dataframe
clean_data.dtypes
```

```python
# Determine unique values in "Month" field
clean_data['Month'].unique()
```

```python
# Noticed a strange value "Order Date" and should be eliminated
clean_data = clean_data[clean_data['Month'] != 'Order Date']

# Cast data types for 3 fields 'Quantity Ordered', 'Price Each' and 'Month'
clean_data['Month'] = clean_data['Month'].astype(int)
clean_data['Quantity Ordered'] = clean_data['Quantity Ordered'].astype(int)
clean_data['Price Each'] = clean_data['Price Each'].astype(float)
```

```python
# Create a new field containing sales value
clean_data['Sales'] = clean_data['Price Each']*clean_data['Quantity Ordered']

# Cast Data type 'float' for 'Sales' field 
clean_data['Sales'] = clean_data['Sales'].astype(float)
```

```python
# Identify total sales by month
result = clean_data.groupby(['Month'])['Sales'].sum()

# Visualization
result.plot(kind = 'bar', figsize=(8, 6))
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Total sales by month')
plt.xticks(rotation = 0, fontsize = 8)

plt.show()
```

### 2. Identify the number of orders by city

```python
# Show values in 'Purchase Address' field
clean_data["Purchase Address"].head()
```

```python
# Create a funtion that returns City
def return_city(address):
    return address.split(',')[1]

# Create a new field containing the city values
clean_data['City'] = clean_data['Purchase Address'].apply(return_city)

clean_data['City'].head()
```

```python
# Identify the number of orders by city
result = clean_data.groupby(['City'])['Order ID'].count()
pie_plot = result.plot(kind='pie', autopct = '%1.0f%%', figsize=(10, 6))
pie_plot.set_ylabel("")
plt.title('The number of order by city')

plt.show()
```

### 3. Identify sales and the number of orders by each product

```python
clean_data.head()
```

```python
# Identify sales and the number of orders by each product
result = clean_data.groupby(['Product']).agg({"Quantity Ordered":"sum", "Sales":"sum"}).reset_index()
result.head()
```

```python
# Visualization
fig , ax1 = plt.subplots(figsize=(8, 6))
ax2 = ax1.twinx() 

ax1.bar(result['Product'] , result['Quantity Ordered'])
ax2.plot(result['Product'] , result['Sales'], color = 'orange' )

ax1.set_xticklabels(result['Product'], rotation = 90, fontsize = 8)

ax1.set_xlabel('Products')
ax1.set_ylabel('Orders')
ax2.set_ylabel('Sales')
ax1.set_title('Sales and orders by each product')
```

### 4. Identify the trend of the top 3 products with the largest sales

```python
# Sort sales in descending order
sort_result = result.sort_values(by = 'Sales', ascending = False)
sort_result.head(3)
```

```python
# Choose 3 products with the highest sales
selected_products = clean_data[clean_data["Product"].isin(["Macbook Pro Laptop", "iPhone", "ThinkPad Laptop"])]

# The size() function counts the number of records in each group generated from groupby() function
# The unstack() function turns the values in the "Product" field from the row to the column --> Creating a pivot table
pivot = selected_products.groupby(['Month', 'Product']).size().unstack()

# Visualization
pivot.plot(kind = 'line', figsize = (8, 6))
plt.xlabel('Month')
plt.ylabel('Order')
plt.title('Total orders by month')
```

### 5. Identify cross-sell products 

#### a) Identify top 5 cross-sell products 

```python
# Show dataframe clean_data
clean_data.head()
```

```python
# Identify records with the same order ID
df_duplicated = clean_data[clean_data['Order ID'].duplicated(keep = False)]
df_duplicated.head()
```

```python
# Concate products into a single cross-sell product
duplicate_products = df_duplicated.groupby(['Order ID'])['Product'].apply(lambda x : ','.join(x))
duplicate_products = duplicate_products.reset_index().rename(columns = {'Product':'Cross-sell'})

duplicate_products.head()
```

```python
# Combine dataframe duplicate_products with dataframe df_duplicated with same Order ID
Cross_sell = df_duplicated.merge(duplicate_products, how = 'left' , on = 'Order ID')

Cross_sell.head()
```

```python
# Eliminate duplicate Order ID values
Cross_sell = Cross_sell.drop_duplicates(subset = ['Order ID']) 

Cross_sell.head()
```

```python
# Visualize top 5 cross-sell products
pie_plot = Cross_sell['Cross-sell'].value_counts()[0:5].plot(kind = 'pie', autopct = '%1.0f%%')
pie_plot.set_ylabel("")
```

#### b) Identify top 5 sales of cross-sell products 

```python
# Identify total sales and the number of orders by Cross-sell
sum_by_product = Cross_sell.groupby('Cross-sell').agg({ "Order ID": "count", "Sales":"sum"}).reset_index()
sum_by_product.head()
```

```python
# Choose top 5 Cross-sell products with largest Orders
sort_result = sum_by_product.sort_values(by = 'Order ID', ascending = False).reset_index().head(5)
sort_result
```

```python
# Visualization
plt.figure(figsize=(8, 6))
plt.bar(sort_result['Cross-sell'], sort_result['Sales'])
plt.xlabel('Cross-sell')
plt.ylabel('Sales')
plt.title('Sales of top 5 cross-sell products')
plt.xticks(rotation = 10, fontsize = 8)

plt.show()
```
